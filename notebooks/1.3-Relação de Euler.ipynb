# Relação de Euler

A partir das expansões de Taylor das funções seno e cosseno, pode-se verificar a relação de Euler:

\( \cos \theta + j \sin \theta = e^{j\theta} \)

#### Importar as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Cria função para mostrar plot

```python
def plot_graph(x, y, title, xlabel, ylabel):
    """
    Plota um gráfico genérico.

    Args:
        x (array-like): Dados do eixo x.
        y (array-like): Dados do eixo y.
        title (str): Título do gráfico.
        xlabel (str): Rótulo do eixo x.
        ylabel (str): Rótulo do eixo y.
    """
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
```

```python
# Define o intervalo de valores para teta
theta = np.linspace(0, 2 * np.pi, 100)

# Calcula o cosseno de teta
cos_theta = np.cos(theta)

# Plota o gráfico
plot_graph(theta, cos_theta, 'Gráfico do Cosseno de Theta', 'Theta (radianos)', 'Cos(Theta)')
```

```python
# Plota o gráfico de -theta versus cos_theta
plot_graph(-theta, cos_theta, 'Gráfico de -Theta vs Cos(Theta)', '-Theta (radianos)', 'Cos(Theta)')
```

```python
# Calcula o seno de teta
sin_theta = np.sin(theta)

# Plota o gráfico
plot_graph(theta, sin_theta, 'Gráfico do Seno de Theta', 'Theta (radianos)', 'Sin(Theta)')
```

```python
# Plota o gráfico
plot_graph(-theta, sin_theta, 'Gráfico do Seno de Theta', 'Theta (radianos)', 'Sin(Theta)')
```
